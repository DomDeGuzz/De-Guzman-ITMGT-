```python
import requests
import time
import matplotlib.pyplot as plt

latitudes = []
longitudes = []
timestamps = []

for _ in range(100):
    response = requests.get("https://api.wheretheiss.at/v1/satellites/25544")
    data = response.json()

    lat = data['latitude']
    lon = data['longitude']
    ts = data['timestamp']

    latitudes.append(lat)
    longitudes.append(lon)
    timestamps.append(ts)

    print(f"Timestamp: {ts}, Lat: {lat}, Lon: {lon}")
    time.sleep(10)  
```

```python
plt.figure(figsize=(20, 10))
plt.plot(longitudes, latitudes, marker='o', linestyle='-', color='yellow')
plt.title("Path of the International Space Station (ISS)")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid(True)
plt.show()
```
